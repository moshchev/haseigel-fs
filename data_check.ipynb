```python
import pandas as pd
from pprint import pp
from IPython.display import display, HTML
```

```python
data10 = pd.read_json('data/test_10.json')
data10
```

```python
data500 = pd.read_json('data/test_500.json')
data500.head(10)
```

```python
data = pd.concat((data10, data500))
data.reset_index(drop=True, inplace=True)
```

```python
from dotenv import load_dotenv
assert load_dotenv()
```

```python
# bring data to db
import os
from sqlalchemy import create_engine

host = os.getenv('DB_HOST')
database = os.getenv('DB_NAME')
user = os.getenv('DB_USER')
password = os.getenv('DB_PASS')
port = os.getenv('DB_PORT')

connection_string  = f'postgresql://{user}:{password}@{host}:{port}/{database}'
engine =  create_engine(connection_string)

data.to_sql('html_data', engine, 'public', if_exists='append', index=False)
```

```python
# get the data
df = pd.read_sql('select * from html_data limit 10', engine)
df.head()
```
